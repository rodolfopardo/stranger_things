```python
#Que cantidad de capítulos han sido analizados 

pd.read_sql('Select count(*) from Stranger', sql)
```

```python
#Cantidad de capitulos por año 

pd.read_sql("SELECT year as 'Años', count(Year) as 'Cantidad de episodios' FROM Stranger GROUP BY Year ",sql)
```

```python
#Promedio de duracion de los capitulos por año 

pd.read_sql("Select year as 'Años', AVG(Runtime) as 'Promedio de duración por minutos' From Stranger GROUP BY Year",sql)
```

1. Podemos observar que las series tienden a durar más a medida que pasan los años. 
2. Este 2019 tiene los minutos más largos.

```python
#Suma de minutos en total de capitulos

pd.read_sql("Select year as 'Años', SUM(Runtime) as 'Duración total por temporada' from Stranger Group by Year", sql)
```

1. En total, 2017 fue la temporada más larga de Stranger Things. 
2. En total, 2016 fue la temporada más baja de Stranger Things.

```python
#Promedio de ratings en total de capitulos 

pd.read_sql("Select year as 'Años', AVG(Rat) as 'Rating promedio' from Stranger GROUP BY Year ORDER BY year", sql)
```

1. En total, 2016 fue la temporada con rating más alto
2. En total, 2017 fue la temporada con rating más bajo
